## 2. Data Preparation

```python
# Removing all non-essential columns: ARREST_KEY, PD_CD, KY_CD, LAW_CODE, JURISDICTION_CODE, X_COORD_CD, Y_COORD_CD, Latitude, Longitude, New Georeferenced Column:
df = df.drop(['ARREST_KEY', 'PD_CD', 'KY_CD', 'LAW_CODE', 'JURISDICTION_CODE', 'X_COORD_CD', 'Y_COORD_CD', 'Latitude', 'Longitude', 'New Georeferenced Column'], axis = 1)
```

```python
# Updated dataset with only relevant information:
df.head(10) 
```

```python
# Replacing Letters in ARREST_BORO column with Borough Names:
df['ARREST_BORO'].replace('K', 'Brooklyn', inplace=True)
df['ARREST_BORO'].replace('M', 'Manhattan', inplace=True)
df['ARREST_BORO'].replace('Q', 'Queens', inplace=True)
df['ARREST_BORO'].replace('S', 'Staten Island', inplace=True)
df['ARREST_BORO'].replace('B', 'Bronx', inplace=True)
df.head(10)
```

```python
# Changing name of LAW_CAT_CD column to make it more intutiive:
df.rename(columns={'LAW_CAT_CD': 'LEVEL_OF_OFFENSE'}, inplace=True)
df.head()
```

```python
# Replacing Letters in LEVEL_OF_OFFENSE column with Level of Offenses:
df['LEVEL_OF_OFFENSE'].replace('F', 'Felony', inplace=True)
df['LEVEL_OF_OFFENSE'].replace('M', 'Misdemeanor', inplace=True)
df['LEVEL_OF_OFFENSE'].replace('V', 'Violation', inplace=True)
df.head()
```

```python
# Changing name of LAW_CAT_CD column to make it more intutiive:
df.rename(columns={'LAW_CAT_CD': 'LEVEL_OF_OFFENSE'}, inplace=True)
df.head()
```

```python
# Replacing Letters in LEVEL_OF_OFFENSE column with Level of Offenses:
df['LEVEL_OF_OFFENSE'].replace('F', 'Felony', inplace=True)
df['LEVEL_OF_OFFENSE'].replace('M', 'Misdemeanor', inplace=True)
df['LEVEL_OF_OFFENSE'].replace('V', 'Violation', inplace=True)
df.head()
```

```python
# Replacing Letters in PERP_SEX column with Male/Female:
df['PERP_SEX'].replace('F', 'Female', inplace=True)
df['PERP_SEX'].replace('M', 'Male', inplace=True)
df.head()
```

```python
# Dropping rows where LEVEL_OF_OFFENSE is in the list of unwanted values:
unwanted_values = ['9', 'I', '(null)']  
df = df[~df['LEVEL_OF_OFFENSE'].isin(unwanted_values)]
```

```python
print('Finding Null Values:')
df.isnull().sum()
```

```python
# Displaying Rows with any null values:
df[df.isnull().any(axis=1)]
```

```python
# Dropping Rows with any null values:
df.dropna(inplace=True)
df
```

```python
# Double checking no null values are present:
df.isnull().sum()
```

## 3. Exploratory Data Analysis (EDA)

### Question 1: Most prevalent gender to commit a crime.

#### (a) Comparing amount of crimes committed by Males versus Females:

```python
# Finding gender that commits most crimes:
gender_counts = df['PERP_SEX'].value_counts()
most_prevalent_gender = gender_counts.idxmax()
most_prevalent_count = gender_counts.max()
print(f"The most prevalent gender to commit a crime is: {most_prevalent_gender} with {most_prevalent_count} occurrences.")
```

```python
# Finding gender that commits least crimes:
gender_counts = df['PERP_SEX'].value_counts()
least_prevalent_gender = gender_counts.idxmin()
least_prevalent_count = gender_counts.min()
print(f"The least prevalent gender to commit a crime is: {least_prevalent_gender} with {least_prevalent_count} occurrences.")
```

```python
# Now lets compare male to female crime count:
gender_counts.plot(kind='bar')
plt.show()
```

#### (b) Exploring level of offense Commited by Males (most prevalent gender to commit a crime):

```python
# Most common crimes committed by males:
male_crimes = df[df['PERP_SEX'] == 'Male']
male_crime_counts = male_crimes['LEVEL_OF_OFFENSE'].value_counts()
print(male_crime_counts.head())
male_crime_counts.head().plot(kind='bar')
plt.show()
```

```python
# Most common crimes committed by males:
female_crimes = df[df['PERP_SEX'] == 'Female']
female_crime_counts = female_crimes['LEVEL_OF_OFFENSE'].value_counts()
print(female_crime_counts.head())
female_crime_counts.head().plot(kind='bar')
plt.show()
```

#### (c) Exploring what borough is most impacted depending on gender:

```python
data = df.groupby(['PERP_SEX', 'ARREST_BORO']).size().reset_index(name='count')
plt.figure(figsize=(10, 6))
sns.barplot(x='ARREST_BORO', y='count', hue='PERP_SEX', data=data, palette='viridis')
plt.show()
```

### Question 2: Months of the year that crimes most take place in comparison to the least.

#### (a) Finding month with most crimes:

```python
# Making a new column called Arrest_Month from Arrest_Date then only took the only from the date
df['ARREST_MONTH'] = df['ARREST_DATE'].str[:2]
df.head()
# Counted the values and then sorted the data from highest to lowest
df.ARREST_MONTH.value_counts().sort_values(ascending = False)
# Ploted
df.ARREST_MONTH.value_counts().sort_values(ascending = False).plot(kind ='bar')
plt.title('Comparison of Months with the Most and Least Crime Incidents')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
df.ARREST_MONTH.value_counts().sort_values(ascending = False)
```

#### (b) Seprated by month, finding out which age groups committed the most crimes:

```python
# Used the groupby function to see the relationship between Arrest Month and the Age group
df_groupedage = df.groupby(['ARREST_MONTH', 'AGE_GROUP']).size().unstack()
# Then using the plot fuction to plot a bar chart 
df_groupedage.plot(kind='bar', stacked=True)
plt.ylabel('Number of Crimes')
plt.title('Monthly Crime Incidents by Age Group')
```

#### (c) In August finding out how crime incidents vary by offense type and age group:

```python
# Created a filter to only get the month of August 
age_df = df[df['ARREST_MONTH']== '08']
# Used the groupby fuction to see the relationship between level of offense and age group
df_groupedage = age_df.groupby(['LEVEL_OF_OFFENSE', 'AGE_GROUP']).size().unstack()
# Ploted this in a bar chart 
df_groupedage.plot(kind='bar', stacked=True)
plt.ylabel('Number of Crimes')
plt.title('Distribution of Crimes in August by Offense Type and Age Group')
```

### Question 3: Boroughs that have highest VS lowest crime rate. Use the population of each borough to count crime rate (count of crimes/populations)

#### (a) Calculating Proportions of Arrest by Borough:

```python
# Calculating Proportions of Arrest by Borough:
BORO_POP = {'Brooklyn': 2561225, 'Bronx': 1356476, 'Manhattan': 1597451, 'Queens': 2252196, 'Staten Island': 490687}
df['BORO_POPULATION'] = df['ARREST_BORO'].map(BORO_POP)
```

```python
bk = (df['ARREST_BORO'].value_counts()['Brooklyn'])/(BORO_POP['Brooklyn'])*100
bx = (df['ARREST_BORO'].value_counts()['Bronx'])/(BORO_POP['Bronx'])*100
si = (df['ARREST_BORO'].value_counts()['Staten Island'])/(BORO_POP['Staten Island'])*100
ma = (df['ARREST_BORO'].value_counts()['Manhattan'])/(BORO_POP['Manhattan'])*100
qn = (df['ARREST_BORO'].value_counts()['Queens'])/(BORO_POP['Queens'])*100
```

```python
boroughs = ['Brooklyn', 'Bronx', 'Staten Island', 'Manhattan', 'Queens']
arrest_proportions = [bk, bx, si, ma, qn]
arrest_proportions
```

```python
plt.bar(boroughs, arrest_proportions, color=['orange', 'red', 'green', 'purple', 'blue'])
plt.xlabel('Borough')
plt.ylabel('Arrest Proportion %')
plt.title('Arrest Proportion by Borough')

plt.show()
```

#### (b) Proportion of crimes committed in each borough by race

```python
df.head()
```

```python
prop = df.groupby(['PERP_RACE', 'ARREST_BORO']).size().reset_index(name='count')
```

```python
prop['population'] = prop['ARREST_BORO'].map(BORO_POP)
prop['proportion (%)'] = prop['count'] / prop['population']*100
prop.sort_values(by='proportion (%)', ascending=False)
```

```python
plot_data = prop.pivot(index='ARREST_BORO', columns='PERP_RACE', values='proportion (%)')

plot_data.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Proportion of Each Race by Borough')
plt.xlabel('Borough')
plt.ylabel('Proportion of Population (%)')
plt.show()
```

```python
# Displaying the information above alternatively using Seaborn: 
male_crimes = df[df['PERP_SEX'] == 'Male']
male_data = male_crimes.groupby(['AGE_GROUP', 'LEVEL_OF_OFFENSE']).size().reset_index(name='count')

female_crimes = df[df['PERP_SEX'] == 'Female']
female_data = female_crimes.groupby(['AGE_GROUP', 'LEVEL_OF_OFFENSE']).size().reset_index(name='count')

fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Male crimes plot
sns.barplot(
    x='AGE_GROUP',
    y='count',
    hue='LEVEL_OF_OFFENSE',
    data=male_data,
    ax=axes[0],
    ci=None
)
axes[0].set_title('Male Crimes by Age Group and Level of Offense')

# Female crimes plot
sns.barplot(
    x='AGE_GROUP',
    y='count',
    hue='LEVEL_OF_OFFENSE',
    data=female_data,
    ax=axes[1],
    ci=None
)

axes[1].set_title('Female Crimes by Age Group and Level of Offense')

plt.show()
```
